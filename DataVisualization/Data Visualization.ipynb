```python
# DATA VISUALIZATION : GRAPHICAL REPRESENTATION OF DATA
# EDA(EXPLORATORY DATA ANALYSIS) : IS AN APPROACH TO ANALYZING DATA SETS 
#                                  TO SUMMARIZE THEIR MAIN CHARACTERISTICS, 
#                                  OFTEN WITH VISUAL METHODS


# SEABORN LIBRARY: UPDATED VERSION OF MATPLOTLIB LIBRARY, 
#                  ATTRACTIVE GRAPHS
```

```python
#METHODS

# HISTOGRAM : A GRAPHICAL DISPLAY OF DATA USING BARS OF DIFFERENT HEIGHT
#             IT'S USED TO CHECK DISTRIBUTION OF CONTINUOUS DATA. 
#             NO DIFFERENCE BETWEEN BINS OF GRAPH

# SCATTER PLOT : CHECK THE RELATIONSHIP OF TWO VARIABLES, POSITIVE REL,
#                NEGATIVE REL

# BAR GRAPH : CHECK THE DISTRUBUTION OF CATEGORICAL VARIABLE/ 
#             DISCRETE VARIABLE. SPACE IS BETWEEN THE BARS

# LINE PLOT : TIME SERIES , TREND, FLUCTUATIONS(변동)

# BOXPLOT : TO CHECK THE OUTLIER IN THE DATASET

# KDE(KERNEL DENSITY ESTIMATE) PLOT: DISTRIBUTION OF SINGLE VARIABLE
#                                    VISUALIZING THE PROBABILITY DENSITY 
#                                    OF A CONTINUOUS VARIABLE
#                                    NON-PARAMETRIC 

# PAIR PLOT : DISTRIBUTION OF MORE THAN ONE VARIABLE

# RELATION PLOT(REL PLOT) : RELATIONSHIP OF VARIABLES (KIND)

# HEATMAP : CORRELATION OF VARIABLES , CORR() , (ANNOT = TRUE)

# HUE : TO GET COLOUR

# MULTICOLLINERITY : WHEN ONE INDEPENDENT VARIABLE IS HIGHLY DEPENDENT 
#                    ON ANOTHER INDEPENDENT VARIABLE

# BUKEH : A PYTHON LIBRARY FOR INTERACTIVE VISUALIZATION THAT 
#         TARGETS WEB BROWSERS FOR REPRESENTATION.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

house_data = pd.read_excel("/users/sanghyunryu/desktop/Trainning/DataVisualization/house_price.xlsx")
print(house_data)
```

```python
# matplot Library
# HISTOGRAM

# ADD LABEL TO DIAGRAM
plt.figure(figsize = (22,7))
plt.hist(house_data["INCOME"])
plt.title("Distribution_of_income_among_different_cities")
plt.xlabel("INCOME")
plt.ylabel("Income_values")
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Histogram.png")
plt.show()
```

```python
# BAR GRAPH

table = pd.pivot_table(data = house_data, index = "CITY",values = "DIST_HOSPITAL", aggfunc = np.sum)
print(table)
plt.figure(figsize = (21,6))

plt.bar(table.index,table["DIST_HOSPITAL"])
plt.xticks(rotation = 70)
plt.title("Distance To Hospital In Different Cities")
plt.xlabel("CITY")
plt.ylabel("DIST_HOSPITAL(M)")
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Bar Graph.png")
plt.show()
```

```python
# SCATTERPLOT

plt.scatter(data = house_data, x = "CITY", y = "RENT")
plt.title("Relationship Between Cities and Rent")
plt.xlabel("City")
plt.ylabel("Rent")
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Scatterplot.png")
plt.show()
```

```python
# LINE PLOT

plt.figure(figsize = (40, 6))
plt.plot(house_data["INCOME"])
plt.title("INCOMES By NUMBER")
plt.xlabel("Number")
plt.ylabel("Income")
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Lineplot.png")
plt.show()
```

```python
# SEABORN LIBRARY

# BARPLOT

sns.countplot(house_data["CITY"])
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Seaborn_BarGraph.png")
plt.show()
```

```python
# HISTOGRAM+KDE

sns.distplot(house_data["RENT"])
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Seaborn_Histogram+KDE.png")
plt.show()
```

```python
# BOXPLOT

sns.boxplot(house_data["INCOME"])
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Seaborn_Boxplot.png")
plt.show()
```

```python
# SCATTERPLOT

sns.scatterplot(house_data["RENT"],house_data["DIST_HOSPITAL"])
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Seaborn_Scatterplot.png")
plt.show()
```

```python
# RELATION PLOT

sns.relplot(x = "RENT", y = "CITY",data = house_data, kind = "scatter", hue = "WIFI")
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Seaborn_Relplot.png")
plt.show()
```

```python
# HEAT MAP : Correlations

sns.heatmap(house_data.corr(), annot = True)
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Seaborn_Heatmap.png")
plt.show()
```

```python
# KDE PLOT : KERNEL DENSITY PLOT

sns.FacetGrid(house_data, hue = "CITY", size = 7).map(sns.kdeplot, "BEDROOM").add_legend()
plt.savefig("/users/sanghyunryu/desktop/Trainning/DataVisualization/Seaborn_KDE.png")
plt.show()
```
