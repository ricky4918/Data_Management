```python
#SUpervised Learning
# It is the learning where the value or result that we want to predict is within the training data (labeled data) 
# and the value which is in data that we want to study is known as Target or Dependent Variable or Response Variable.
# All the other columns in the dataset are known as the Feature or Predictor Variable or Independent Variable.

#Supervised Learning is classified into two categories:
# Clarification: Here our target variable consists of the categories.
# Regression: Here our target variable is continuous and we usually try to find out the line of the curve.
```

```python
#K-nearest neighbor : the knn algorithm assumes that similar things exist in close proximity
#                     This algorithm is used to solve the classification model problems

#The Knn algorithm
# 1. load the data
# 2. initialize k to your chosen number of neighbors
# 3. for each example in the data
# 3.1 measure the distance between new data point and old data points(training dataset)
# 3.2 add the distance and the index of the example to an ordered collection
# 4. sort the ordered collection of distances and indices from smallest to largest (in ascending order) by the distances
# 5. pick the first k entries from the sorted collection
# 6. get the labels of the selected k entries
# 7. if regression, return the mean of the k labels
# 8. if classification, return the mode of the k labels

# larger k value means smother curves of separation resulting in less complex models.
# Whereas, smaller k value tends to overfit the data and resulting in complex models.


# How to decide k(odd number)

# Measure the distance between new data point and old data points(training dataset)
# Sort the output in ascending order
# Select top rows of asecending order
# Most frequent class/category
# Based on majority of votes select class/category
```

```python
# EUCLIDEAN DISTANCE
import math
# Example points in 3-dimensional space...
x = (5, 6, 7)
y = (8, 9, 9)
distance = math.sqrt(sum([(a - b) ** 2 for a, b in zip(x, y)]))
print("Euclidean distance from x to y: ",distance)
```

```python
# MANHATTAN DISTANCE
def ManhattanDistance(x, y, n):
    distance = 0
    n = 3

    for i in range(n):
        distance += (abs(x[i]-y[i]))
    return distance

x = [-1, 1, 3]
y = [5, 6, 5]


print(ManhattanDistance(x,y,n))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Assign colum names to the dataset
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']

# Read dataset to pandas dataframe
dataset = pd.read_csv(url, names=names)
print(dataset)
```

```python
#lable colum and row to x-axis and y-axis
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
#The above script splits the dataset into 80% train data and 20% test data.
```

```python
#KNN Algorithm
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
